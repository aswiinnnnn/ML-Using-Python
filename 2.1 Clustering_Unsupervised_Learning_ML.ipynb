# **KMeans**

```python
sub_one = [5, 4, 13, 11, 13, 5, 3, 4, 3, 10, 12, 13, 14, 12, 11, 6, 7, 6, 5, 11, 14, 10]
sub_two = [15, 18, 22, 21, 21, 19, 15, 16, 16, 21, 23, 22, 22, 20, 22, 17, 18, 14, 17, 24, 22, 23]
```

```python
import matplotlib.pyplot as plt

plt.scatter(sub_one, sub_two)
```

```python
data = list(zip(sub_one, sub_two))
data
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, n_init='auto')
kmeans.fit(data)

c = kmeans.labels_
print(c)
```

```python
plt.scatter(sub_one, sub_two, c=kmeans.labels_)
plt.show()
```

# **Quality of the cluster**



>The Silhouette Coefficient is a metric used to evaluate the quality of clustering in unsupervised learning. It measures how well each data point fits into its assigned cluster compared to other clusters.





 Silhouette Coefficient is a value between -1 and 1:



*   The point is well-clustered and far from neighboring clusters.
*   0: The point is on or near the boundary between two clusters.
*   -1: The point is likely assigned to the wrong cluster.








```python
from sklearn import metrics

metrics.silhouette_score(data, c, metric ='euclidean')
```

# **Elbow Method**
>The Elbow Method helps you find the best number of clusters for your data.

You try different numbers of clusters (e.g., 1, 2, 3, etc.).

*   You check how well the data fits into each number of clusters.
*   You make a graph of the results.
*   The "elbow" (bend) of the graph shows the best number of clusters.

```python
inertias=[]
for i in range (2,11):
  kmeans = KMeans(n_clusters = i, n_init= 'auto')
  kmeans.fit(data)
  inertias.append(kmeans.inertia_)

print(inertias)
plt.plot(range(2,11),inertias,marker='o')
plt.title("Elbow method")
plt.xlabel("No of Clusters")
plt.ylabel("inertias")
```

```python
inertias=[]
for i in range (2,11):
  kmeans = KMeans(n_clusters = i, n_init= 'auto')
  kmeans.fit(data)
  c=kmeans.labels_
  inertias.append(metrics.silhouette_score(data, c, metric ='euclidean'))

print(inertias)
plt.plot(range(2,11),inertias,marker='o')
plt.title("Elbow method")
plt.xlabel("No of Clusters")
plt.ylabel("inertias")
```
